```python
# reading the riders file 2020
q2_2020 = pd.read_csv('/content/drive/MyDrive/Divvy_Trips_2020_Q1 - Divvy_Trips_2020_Q1.csv')
```

```python
# renaming some columns in 2019 dataframe to match 2020
q1_2019 =q1_2019.rename(columns={'trip_id':'ride_id','bikeid':'rideable_type','start_time':'started_at','end_time':'ended_at','from_station_name':'start_station_name','from_station_id':'start_station_id','to_station_name':'end_station_name','to_station_id':'end_station_id','usertype':'member_casual'})
```

```python
# concatenating the two data frame, 2019,2020
df_all= pd.concat([q1_2019,q2_2020])
df_all
```

```python
# dropping some columns
df_all= df_all.drop(['start_lat', 'start_lng', 'end_lat', 'end_lng','gender', 'birthyear','tripduration'], axis = 1)
```

```python
# looking throught the new dataframe
df_all
```

```python
# replacing Subscriber with member and customer with casual in member_casual columns
mapping = {'Subscriber': 'member','casual':'casual', 'Customer':'casual'}
df_all['member_casual'] = df_all['member_casual'].replace(mapping)
df_all
```

```python
# converting columns to datetime
df_all['started_at'] = pd.to_datetime(df_all['started_at'], format='%Y-%m-%d %H:%M:%S')
df_all['ended_at']= pd.to_datetime(df_all['ended_at'], format='%Y-%m-%d %H:%M:%S')
```

```python
# calculating ridelength
df_all['ride_length'] = df_all['ended_at'] - df_all['started_at']
df_all
```

```python
# change to seconds
df_all['ride_length'] = pd.to_timedelta(df_all['ride_length']).dt.total_seconds()
df_all
```

```python
# datetime
df_all['started_at'] = pd.to_datetime(df_all['started_at'])
```

```python
# converting to day of the week
df_all['days of_the_week'] = df_all['started_at'].dt.day_name()
df_all
```

```python
# eliminating all negative entries in ride_length
df_trip = df_all[df_all['ride_length'] > 0]
df_trip
```

```python
# eliminating all test ride
df_tripfinal = df_trip[df_trip['start_station_name']!='HQ QR']
df_tripfinal
```

```python
# using aggregation
#df_tripfinal['ride_length'].describe()
df_tripfinal['ride_length'].agg({'mean':'mean','median':'median','max':'max','min':'min'})
```

```python
# Compare casual vs member
df_tripfinal.groupby('member_casual')['ride_length'].agg(['mean','median','max','min'])
```

```python
# compare casual vs member by each day of the week
df_casula_vs_member = df_tripfinal.sort_values('days of_the_week', ascending = True).groupby(['member_casual','days of_the_week'])['ride_length'].agg('mean')
#df_casula_vs_member.order(level='days of_the_week')
df_casula_vs_member
```

```python
# Define correct weekday order
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Convert weekday column to ordered categorical type
df_tripfinal["days of_the_week"] = pd.Categorical(df_tripfinal["days of_the_week"], categories=weekday_order, ordered=True)

# Group by weekday and sum values
result = df_tripfinal.groupby(['member_casual','days of_the_week'])['ride_length'].agg('mean')
result
```

```python
g = sns.barplot(data=df_tripfinal, x='days of_the_week', y='ride_length', hue='member_casual', errorbar = None)
plt.xticks(rotation=90)
plt.show()
```
